```python
import pandas as pd
dataset = pd.read_csv("../../Expanded_50_Startups.csv")
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset
```

```python
dataset.columns #view the columns in the dataset
dataset.columns = dataset.columns.str.replace(" ", "_")
dataset
```

```python
#Split input and output data into Different variables
independent = dataset[['R_D_Spend', 'Administration', 'Marketing_Spend', 
       'State_Florida', 'State_New_York']]
dependent = dataset[['Profit']]
```

```python
#split Training and Test datasets
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#!pip install lightgbm
```

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
regressor = lgb.LGBMRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    objective='regression',
    random_state=42
)
regressor.fit(xTrain, yTrain)
```

```python
yPred = regressor.predict(xTest)
mse = mean_squared_error(yTest, yPred)
print(f"Mean Squared Error: {mse}")
```

```python
from sklearn.metrics import r2_score
rScore = r2_score(yTest, yPred)
rScore
```

```python
#Save the finalized model
import pickle
fileName = "Finalized_LGBoost_Model.sav"
pickle.dump(regressor, open(fileName,'wb'))
```
