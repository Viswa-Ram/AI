```python
import pandas as pd #import pandas library
dataset = pd.read_csv("Sales_Data.csv") #read dataset
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True) #Convert categorical data into numerical data and remove first column for repeatation 
dataset = dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns #view the columns in the dataset
```

```python
#Split input and output data
independent = dataset[['Marketing Spend', 'Store Size','State_Florida', 'State_Illinois', 'State_New York', 'State_Texas']]
dependent = dataset['Sales']
```

```python
independent
```

```python
dependent
```

```python
#Split training data and test data
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest = train_test_split(independent, dependent, test_size=1/3, random_state=0)
```

```python
xTrain
```

```python
#Create MLR model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xTrain, yTrain)
```

```python
weight = regressor.coef_ #find the weight values
```

```python
weight
```

```python
bias = regressor.intercept_ #find the bias value
bias
```

```python
yPred = regressor.predict(xTest) #predict with test data
```

```python
#Metrics validation using r2 score function
from sklearn.metrics import r2_score
rScore = r2_score(yTest,yPred)
rScore
```

```python
#Save the finalized model
import pickle
fileName = "Finalized_MLR_Model.sav"
pickle.dump(regressor, open(fileName,'wb'))
```

```python
loadedModel = pickle.load(open("Finalized_MLR_Model.sav","rb")) #reload the model
```
