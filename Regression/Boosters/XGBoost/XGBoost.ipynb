```python
import pandas as pd
dataset = pd.read_csv("../../50_Startups.csv")
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset = dataset.astype(int)
```

```python
dataset.columns #view the columns in the dataset
```

```python
#Split input and output data into Different variables
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
dependent = dataset[['Profit']]
```

```python
#split Training and Test datasets
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
regressor = XGBRegressor(
    n_estimators=100,
    learning_rate=0.3,
    max_depth=3,
    objective='reg:squarederror',
    random_state=42
)

regressor.fit(xTrain, yTrain)
```

```python
yPred = regressor.predict(xTest)
```

```python
from sklearn.metrics import r2_score
rScore = r2_score(yTest, yPred)
rScore
```

```python
#Save the finalized model
import pickle
fileName = "Finalized_XGBoost_Model.sav"
pickle.dump(regressor, open(fileName,'wb'))
```
