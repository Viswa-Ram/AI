```python
import pandas as pd
dataset = pd.read_csv("../insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.astype(int)
```

```python
#split independent and dependenet data
indenependent = dataset[['age','bmi', 'children', 'sex_male',  'smoker_yes']]
dependent = dataset['charges']
                         
```

```python
#split into training and test Data
# from sklearn.model_selection import train_test_split
# xTrain, xTest, yTrain, yTest = train_test_split(indenependent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson'], 'max_features': ['None', 'sqrt', 'log2'], 'n_estimators':[50, 100]}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose = 3, n_jobs=-1)

grid.fit(indenependent, dependent)
```

```python
#print best parameter after tuning
re = grid.cv_results_
#print(re)

# grid_predictions = grid.predict(xTest)

#print classification report
# from sklearn.metrics import r2_score
# rScore = r2_score(yTest, grid_predictions)

print('The R-Score of the best perameter {}:'.format(grid.best_params_))
```
