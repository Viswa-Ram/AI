```python
import pandas as pd
dataset = pd.read_csv("../50_Startups.csv")
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset = dataset.astype(int)
```

```python
dataset.columns #view the columns in the dataset
```

```python
#Split input and output data into Different variables
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent = dataset[['Profit']]
```

```python
#split Training and Test datasets
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#modeil creation
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='absolute_error', splitter="best")
regressor = regressor.fit(xTrain, yTrain)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
yPred = regressor.predict(xTest)
```

```python
from sklearn.metrics import r2_score
rScore = r2_score(yTest, yPred)
```
